# VisualAstro Colormaps Guide

VisualAstro registers `cmasher` colormaps and offers colormap visualization and manipulation tools. Use `get_cmap` to retrieve a colormap. You can optionally set the range of the colormap with `cmap_range` and the color that represents bad values (e.g. `numpy.inf`, `numpy.nan`) with `bad_color`.

```python
import matplotlib.pyplot as plt
import visualastro as va

plt.style.use(['full', 'computer-modern-serif'])
```

```python
va.get_cmap('cosmic')
```

```python
va.get_cmap('cosmic', cmap_range=[.3,1])
```

```python
va.get_cmap('emerald', cmap_range=[.4,1], bad_color='purple')
```

### Create a Colormap From Colors

Create a colormap using a list of colors.

```python
turbo6 = va.get_colors('turbo6')
astro_seq = va.get_colors('astro_seq')
jade_seq = va.get_colors('jade_seq')

va.help.color(turbo6)
va.help.color(astro_seq)
va.help.color(jade_seq)
```

```python
va.create_cmap(turbo6)
```

```python
va.create_cmap(astro_seq)
```

```python
va.create_cmap(jade_seq)
```

### Plot Colormap Lightness

Here are the lightness $L^*$ plots of colormaps, similar to the $L^*$ plots by Matplotlib: [https://matplotlib.org/stable/users/explain/colors/colormaps.html#lightness-of-matplotlib-colormaps](https://matplotlib.org/stable/users/explain/colors/colormaps.html#lightness-of-matplotlib-colormaps).

```python
va.config.figsize = (12,6)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(14,10))
ax[0].set_title('Perceptually Uniform Colormaps')
va.plot_cmap_lightness(['viridis', 'cosmic', 'plasma', 'inferno', 'emerald', 'cividis'], ncols=6, offset=.2, ax=ax[0])
va.plot_cmap_lightness(['turbo', 'BrBG', 'RdPu', 'gothic', 'redshift'], ncols=5, offset=.2, ax=ax[1])
va.plot_cmap_lightness(['rainforest', 'iridescent', 'voltage', 'emerald', 'bubblegum', 'tol.rainbow'], ncols=6, offset=.2, ax=ax[2])
plt.show()
```
